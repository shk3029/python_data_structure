### 문제

- 2×n 크기의 직사각형을 1×2, 2×1 타일로 채우는 방법의 수를 구하는 프로그램을 작성하시오.
- 아래 그림은 2×5 크기의 직사각형을 채운 한 가지 방법의 예이다.


- input : n (1이상 1000이하)
- output : 2xn 크기의 직사각형을 채우는 방법의 수를 10,007로 나눈 나머지

- 2x0 = 0개 f(0)
- 2x1 = 1개 f(1)
- 2x2 = 1+1개 f(2)
- 2x3 = 2+2 f(3) = 2xf(2)
- 2x4 = 8+8 f(4) = 2xf(3)
- 2x5 = 16+16 f(5) = 2x(f4)

```python
def solution(n):
    cache = [0 for i in range(n+1)]
    
    cache[1] = 1

    if n==1:
        return cache[n]
    
    cache[2] = 2
    
    if n==2:
        
        return cache[n]

    for i in range(3,n+1):
        cache[i] = cache[i-2] + cache[i-1]
    
    return cache[n] % 10007

n = input()
solution(int(n))
```

```python
dp = [0]*1001
dp[1]=1
dp[2]=2

for index in range(3, 1001):
    dp[index] = dp[index-1] + dp[index-2]

n = int(input())
print(dp[n] % 10007)
```

### 해설

- 동적계획법은 빠르게 점화식을 찾는게 중요하다
    - 점화식이란, 이웃하는 두개의 항 사이에 성립하는 관계를 나타낸 관계식을 의미함
